```python
import time
import requests 
import pandas as pd
from bs4 import BeautifulSoup as bs
from selenium import webdriver 
from selenium.webdriver.firefox.options import Options
import json
```

```python
url = "https://www.sofascore.com/pt/time/futebol/fluminense/1961"

option = Options()
option.headless = True
driver = webdriver.Firefox()

driver.get(url)

driver.execute_script("window.scrollTo(0, 300);")
time.sleep(5)
driver.execute_script("window.scrollTo(0, 600);")
time.sleep(5)
driver.execute_script("window.scrollTo(0, 1800);")
time.sleep(5)
driver.find_element_by_css_selector("div.TabsWrapper__Wrapper-sc-5ag92o-0:nth-child(2) > div:nth-child(1) > a:nth-child(2)").click()
time.sleep(5)

element = driver.find_element_by_xpath("/html/body/div[1]/main/div/div[2]/div[1]/div[1]/div[5]/div[2]/div[2]/div/table")
html_content = element.get_attribute('outerHTML')
print(html_content)
driver.quit()
```

```python
soup = bs(html_content, 'html.parser')
table = soup.find(name='table')
```

```python
df_full = pd.read_html(str(table))[0]
df_full
```

```python
df_full.to_csv('jogadores.csv',index = False)
```
